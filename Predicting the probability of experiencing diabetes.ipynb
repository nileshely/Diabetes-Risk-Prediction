```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("diabetes_risk_prediction_dataset.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
# Visualization 1: Age distribution
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Age', bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualization 2: Gender distribution
plt.figure(figsize=(6,6))
df['Gender'].value_counts().plot(kind='pie', autopct='%1.2f%%')
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```

```python
# Visualization 3: Class distribution
plt.figure(figsize=(6,6))
df['class'].value_counts().plot(kind='pie', autopct='%1.2f%%')
plt.title('Class Distribution')
plt.ylabel('')
plt.show()
```

```python
# Visualization 4: Correlation heatmap
plt.figure(figsize=(16, 10))
encoded_df = pd.get_dummies(df)
sns.heatmap(encoded_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Visualization 5: Count of diabetes symptoms
symptoms_cols = ['Polyuria', 'Polydipsia', 'sudden weight loss', 'weakness', 'Polyphagia',
                 'Genital thrush', 'visual blurring', 'Itching', 'Irritability',
                 'delayed healing', 'partial paresis', 'muscle stiffness', 'Alopecia']
symptoms_count = df[symptoms_cols].apply(pd.value_counts).transpose()
plt.figure(figsize=(12,6))
symptoms_count.plot(kind='bar', stacked=True)
plt.title('Count of Diabetes Symptoms')
plt.xlabel('Symptom')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(title='Presence')
plt.show()
```

```python
# Visualization 6: Age vs. Obesity
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Obesity', y='Age')
plt.title('Age vs. Obesity')
plt.xlabel('Obesity')
plt.ylabel('Age')
plt.show()
```

```python
# Visualization 7: Age vs. Class
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='class', y='Age')
plt.title('Age vs. Class')
plt.xlabel('Class')
plt.ylabel('Age')
plt.show()
```

```python
# Visualization 8: Gender vs. Class
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Gender', hue='class')
plt.title('Gender vs. Class')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 9: Age distribution by class
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Age', hue='class', bins=20, kde=True)
plt.title('Age Distribution by Class')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend(title='Class')
plt.show()
```

```python
# Visualization 10: Pairplot
sns.pairplot(df, hue='class', diag_kind='kde')
plt.suptitle('Pairplot of Features by Class')
plt.show()
```
